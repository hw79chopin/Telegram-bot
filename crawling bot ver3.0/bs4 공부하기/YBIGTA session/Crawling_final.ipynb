#### HTTP 상태 코드
 - 1xx (정보): 요청을 받았으며 프로세스를 계속한다
 - 2xx (성공): 요청을 성공적으로 받았으며 인식했고 수용하였다
 - 3xx (리다이렉션): 요청 완료를 위해 추가 작업 조치가 필요하다
 - 4xx (클라이언트 오류): 요청의 문법이 잘못되었거나 요청을 처리할 수 없다
 - 5xx (서버 오류): 서버가 명백히 유효한 요청에 대해 충족을 실패했다

[출처: 위키피디아](https://ko.wikipedia.org/wiki/HTTP_%EC%83%81%ED%83%9C_%EC%BD%94%EB%93%9C)

```python
html = '''
<html>
  <head>
    <title>Example HTML</title>
  </head>
  <body>
    <div id='first' class='Example'>
      <h3 title='Good Content Title'>Contents Title</h3>
      <p>Test contents</p>
    </div>
    <div id='second' class='Example'>
      <p>Text1</p>
      <p>Text2</p>
      <p>Text3</p>
    </div>
  </body>
</html>'''
```

### 여러 페이지 크롤링 할 때는?

- url에 page관련 parameter가 있는 경우
- url에 page관련 parameter가 없는 경우

```python
#url에 page관련 parameter가 있는 경우

all_title = []
for i in range(2):
    page = i * 10 +1
    url = 'https://search.naver.com/search.naver?&query=여수%20여행&where=post&start=' + str(page)
    resp = requests.get(url)
    soup = BeautifulSoup(resp.text, 'html.parser')
    
    title = soup.find_all('a', class_='sh_blog_title')
    for item in title:
        all_title.append(item.get_text())
all_title
```

```python
#url에 page관련 parameter가 없는 경우

all_title = []
for i in range(2):
    page = i * 10 +1
    params={'start':page} 
    
    url = 'https://search.naver.com/search.naver?where=post&sm=tab_jum&query=%EC%97%AC%EC%88%98+%EC%97%AC%ED%96%89'
    resp = requests.get(url, params = params)
    soup = BeautifulSoup(resp.text, 'html.parser')
    
    title = soup.find_all('a', class_='sh_blog_title')
    for item in title:
        all_title.append(item.get_text())
all_title
```
